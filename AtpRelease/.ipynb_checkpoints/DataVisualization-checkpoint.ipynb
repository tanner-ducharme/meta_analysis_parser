```python
def sort_dict_by_values(dictionary, min_freq):
    sorted_list = [(k, dictionary[k]) for k in sorted(dictionary, key=dictionary.get, reverse=True)]
    
    
    sorted_dict = {}
    for e in sorted_list[:1000]:
        if e[1] >= min_freq:
            sorted_dict[e[0]] = e[1]
    
    return sorted_dict
```

```python
def top_n_from_dict(dictionary, n):
    ctr = 0
    new_dict = {}
    for key in dictionary.keys():
        if ctr<n:
            new_dict[key] = dictionary[key]
        ctr +=1
    return new_dict
            
```

```python
tally = {}
for term in query_terms:
    tally[term] = 0
    for row in text_df.iterrows():
        row = row[1]
        tally[term]+=row.all_text.count(term)
    
```

```python
def getScatterplotData(df, mechan_words):
    scatterplot_data_R  = []
    scatterplot_data_NR = [] # will be list of tuples (# atp words, # mechanical words, relevance)
    for row in df.iterrows():
        row = row[1]
        atp_count = row.all_text.count('atp') + row.all_text_corpus.count('adenosine triphosphate')

        mechanical_count = 0
        for word in mechan_words:
            mechanical_count += row.all_text.count(word)

        if row.relevant:           
            scatterplot_data_R.append((atp_count, mechanical_count, row.relevant))
        else:
            scatterplot_data_NR.append((atp_count, mechanical_count, row.relevant))
    return (scatterplot_data_R, scatterplot_data_NR)
```

```python
num_top_words = [3,5,10,15,20,50, 100]

for num in num_top_words:
    cur_tally = top_n_from_dict(tally, num+2)
    if 'atp' in cur_tally:
        cur_tally.pop('atp')
    if 'adenosine triphosphate' in cur_tally:
        cur_tally.pop('adenosine triphosphate')

    relevant, irrelevant = getScatterplotData(text_df, cur_tally.keys())
    plt.figure(figsize=(20,10))
    for point in random.sample(relevant , 200) + random.sample(irrelevant , 200):
        
        if point[2]: # relevant == True
            plt.plot(point[0], point[1], '+', mew=2, ms=10,
                    color = 'blue', label = "Relevant")
        else:
            plt.plot(point[0], point[1], 'o',
                    color = 'red', label = "Irrelevant")
       
    plt.xlabel('atp mentions')
    plt.ylabel('mechanical mentions')
    title = "using {} top mechanical words".format(num)
    plt.title(title)
    
    
```
